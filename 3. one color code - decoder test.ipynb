### 載入測試數據

```python
import pandas as pd
import matplotlib.pyplot as plt
df_test = pd.read_csv("s_result_all_100.csv", index_col=0)
X_test = df_test.drop('0',axis='columns')     # 設置輸入變量
Y_test = df_test.loc[:, '0']

```

### 載入模型

```python
import pickle

# load the trained decision tree classifier from Pickle
decision_tree_pkl_filename = 'decision_tree.pkl'
# Open the file to load as pkl file
with open(file=decision_tree_pkl_filename, mode='rb') as f:
    regr=pickle.load(f)
```

### 測試

```python
regr_y_result = regr.predict(X_test.loc[180:360, :])
```

### 分析測試結果

```python
import numpy as np
#print(np.around(regr_y_result))
plt.plot(regr_y_result)
print("np.std=", np.std(regr_y_result[5:174]-Y_test[5:174]))
```
